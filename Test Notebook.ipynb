Once the union was executed, we needed to replace the `NaN` values in the `Infertility` column with `False` to indicate those patients who were not diagnosed with infertility.

```python
PCOS_main['Infertility'] = PCOS_main['Infertility'].replace(np.nan, False)
```

Replace the BMI column with the [correct calculation](https://www.nhlbi.nih.gov/health/educational/lose_wt/BMI/bmicalc.htm). 

```python
PCOS_main['Height (m)'] = PCOS_main['Height(Cm) '] / 100

PCOS_main['BMI'] = PCOS_main['Weight (Kg)'] / (PCOS_main['Height (m)'] ** 2)
```

Replace `FSH/LH` with the correct calculation.

```python
PCOS_main['FSH/LH'] = PCOS_main['FSH(mIU/mL)'] / PCOS_main['LH(mIU/mL)']
```

Replace `Waist:Hip Ratio` with the correct calculation. 

```python
PCOS_main['Waist:Hip Ratio'] = PCOS_main['Waist(inch)'] / PCOS_main['Hip(inch)']
```

Add a column `Height (ft in)` converting `Height cm` to feet and inches for readability. 

```python
#Convert height to feet and inches using Regular Expressions. 
#https://stackoverflow.com/questions/26986655/changing-height-feet-and-inches-to-an-integer-in-python-pandas
```

Add a column `Weight (lbs)` converting weight in kg to weight in lbs for readability.

```python
PCOS_main['Weight (lbs)'] = PCOS_main['Weight (Kg)'] * 2.205
```

Create a column `Blood Type (str)` that is represented by the string value vs. the numerical representation defined. 

```python
PCOS_main['Blood Type (str)'] = PCOS_main['Blood Group'].replace([11, 12, 13, 14, 15, 16, 17, 18], ['A+', 'A-', 'B+', 'B-', 'O+', 'O-', 'AB+', 'AB-'] )
```

Convert blood pressure to the correct format systolic over diastolic. [Reference Link](https://www.heart.org/en/health-topics/high-blood-pressure/understanding-blood-pressure-readings).

```python
PCOS_main['Blood Pressure (str)'] = PCOS_main['BP _Systolic (mmHg)'].astype(str) + '/' + PCOS_main['BP _Diastolic (mmHg)'].astype(str)

PCOS_main['Blood Pressure (str)'] = PCOS_main['Blood Pressure (str)'].replace(['nan/nan'], np.nan)

PCOS_main['Blood Pressure (str)']
```

Correct spelling of column header.

```python
PCOS_main.rename(columns = {'No. of aborptions': 'No. of Abortions'}, inplace = True)
```

 - Change non-string types to floats for numerical variables. 
 - Received `ValueError: Unable to parse string "1.99." at position 123.`
 - Used replace method to `replace` method into format that can be coerced with `to_numeric`.

```python
PCOS_main['Waist:Hip Ratio'] = pd.to_numeric(PCOS_main['Waist:Hip Ratio'])
```

```python
PCOS_main['II    beta-HCG(mIU/mL)'] = PCOS_main['II    beta-HCG(mIU/mL)'].replace(['1.99.'], '1.99')
PCOS_main['II    beta-HCG(mIU/mL)'] = pd.to_numeric(PCOS_main['II    beta-HCG(mIU/mL)'])
```

- For `PCOS_main['AMH(ng/mL)']` recieved `ValueError` at position 307 with the value 'a'. 
- Since this was the only string value in this numerical column, I chose to replace this string with the mean of the column to avoid dropping that record

```python
PCOS_main['AMH(ng/mL)'] = pd.to_numeric(PCOS_main['AMH(ng/mL)'], errors = 'coerce').astype('float64')
PCOS_main['AMH(ng/mL)'] = PCOS_main['AMH(ng/mL)'].fillna(PCOS_main['AMH(ng/mL)'].mean())
```

- Change the type of the categorical variable columns to `bool` from `float64`:
 - `PCOS (Y/N)`
 - `Pregnant (Y/N)`
 - `Weight gain (Y/N)`
 - `hair growth (Y/N)`
 - `Skin darkening (Y/N)`
 - `Hair loss (Y/N)`
 - `Acne (Y/N)`
 - `Fast Food (Y/N)`
 - `Reg.Exercise (Y/N)`

```python
PCOS_main['PCOS (Y/N)'] = PCOS_main['PCOS (Y/N)'].astype('bool')
PCOS_main['Pregnant(Y/N)'] = PCOS_main['Pregnant(Y/N)'].astype('bool')
PCOS_main['Weight gain(Y/N)'] = PCOS_main['Weight gain(Y/N)'].astype('bool')
PCOS_main['hair growth(Y/N)'] = PCOS_main['hair growth(Y/N)'].astype('bool')
PCOS_main['Skin darkening (Y/N)'] = PCOS_main['Skin darkening (Y/N)'].astype('bool')
PCOS_main['Hair loss(Y/N)'] = PCOS_main['Hair loss(Y/N)'].astype('bool')
PCOS_main['Acne (Y/N)'] = PCOS_main['Acne (Y/N)'].astype('bool')
PCOS_main['Reg.Exercise(Y/N)'] = PCOS_main['Reg.Exercise(Y/N)'].astype('bool')
```
